```python
import os
import shutil

source_dir = '/content/drive/MyDrive/text-generation-webui/models/'
dest_dir = '/content/drive/MyDrive/LLMS/'

for filename in os.listdir(source_dir):
    source_file = os.path.join(source_dir, filename)
    dest_file = os.path.join(dest_dir, filename)
    if os.path.exists(dest_file):
        print(f"File {filename} already exists in the destination directory. Skipping...")
        continue
    shutil.copytree(source_file, dest_file)
    print(f"Copied {filename} to the destination directory.")
```
